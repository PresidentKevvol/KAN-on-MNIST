```python
# load with utility code
import mnist_reader
x_tr, y_tr = mnist_reader.load_mnist(join('fashion-mnist-master', 'data', 'fashion'), kind='train')
x_te, y_te = mnist_reader.load_mnist(join('fashion-mnist-master', 'data', 'fashion'), kind='t10k')
```

```python
x_tr = x_tr.reshape((60000, 28, 28))
x_te = x_te.reshape((10000, 28, 28))
```

```python
def show_images(images, title_texts):
    cols = 10
    rows = int(len(images)/cols) + 1
    plt.figure(figsize=(30,20))
    index = 1    
    for x in zip(images, title_texts):        
        image = x[0]        
        title_text = x[1]
        plt.subplot(rows, cols, index)        
        plt.imshow(image, cmap=plt.cm.gray)
        if (title_text != ''):
            plt.title(title_text, fontsize = 15);        
        index += 1
```

```python
items_grouped = []
for i in range(0, 10):
    cur_items = []
    j = 0
    while True:
        if y_tr[j] == i:
            cur_items.append(j)
        j += 1
        if len(cur_items) >= 5:
            break
    items_grouped.append(cur_items)

items_grouped = np.array(items_grouped)
items_grouped = items_grouped.T.reshape((50,))
```

```python
images_2_show = []
titles_2_show = []



for i in range(0, 50):
    r = items_grouped[i]
    images_2_show.append(x_tr[r])
    titles_2_show.append(str(r))    
    # titles_2_show.append('training image [' + str(r) + '] = ' + str(y_tr[r]))    

# for i in range(0, 10):
#     r = random.randint(1, 10000)
#     images_2_show.append(x_te[r])        
#     titles_2_show.append('test image [' + str(r) + '] = ' + str(y_te[r]))    

show_images(images_2_show, titles_2_show)
```
